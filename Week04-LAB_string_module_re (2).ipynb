1. Ask user for an input of his/her password and store it in a variable called password
Print the password length first then create function validating if the password has more than eight characters. If the password fulfills the above condition and has more then eight character print 'valid password' otherwise print invalid password and ask the user to enter the password again.

````python
# aks the user to enter password
password = input("Enter your password: ")

# prints the password length 
print("Password length:", len(password))

# Define a function to check if the password is valid
def check_password(pw):
    if len(pw) > 8:
        print("Valid password")
    else:
        print("Invalid password")
        new_pw = input("Enter your password again: ")
        check_password(new_pw)

# Call the function 
check_password(password)

````

3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python

def convert_password(password):
    result = ""
    for letter in password:
        ascii_value = ord(letter)    
        ascii_char = str(ascii_value)   
        result = result + ascii_char    
    return result


pw = input("Enter your password: ")


ascii_pw = convert_password(pw)
print("Your password in ASCII:", ascii_pw)

````

4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
def check_password(p):
    upper = False
    lower = False
    digit = False
    special = False

    if " " in p:
        print("Password should not contain spaces")
        return

    for c in p:
        if c.isupper():
            upper = True
        if c.islower():
            lower = True
        if c.isdigit():
            digit = True
        if c == "@":
            special = True

    if upper and lower and digit and special:
        print(True)
    else:
        print(False)

pw = input("Enter password: ")
check_password(pw)
````
